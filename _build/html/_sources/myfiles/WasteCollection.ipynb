# Project Specifications Document for Waste Collection Optimization in Guelma

## 1. Project Overview

In Guelma, the management of waste collection trucks is currently inefficient, with trucks driving through neighborhoods without specific information on bin locations.
Recently, Guelma has implemented a new waste management system in some selected streets, where large communal dumpsters have been removed. Instead, residents in these areas place their bins outside their doors for pickup. This project focuses specifically on supporting this new system, aiming to develop a mobile application that will improve collection efficiency in these designated streets. The application will enhance waste collection logistics for garbage trucks while providing a better experience for residents in the affected areas.

## 2. Project Objectives

The primary goal of this project is to create an application accessible by both garbage collectors and citizens, with the following objectives:

- Citizen Reporting: Allow citizens to notify collectors when their bins are set out **during non-scheduled collection times**.
- Real-Time Updates: Enable collectors to see the real-time locations of bins on a map of Guelma.
- Route Optimization: Automatically generate the most efficient route for the truck to collect all bins, ensuring a systematic and time-saving process.

## 3. Functional Requirements

### 3.1 User Types and Roles

- Citizens: Individuals who place bins outside for collection and use the app to signal the availability and location.
- Garbage Collectors: Waste collection personnel who access the app to view bin locations and follow optimized routes.

### 3.2 Citizen Features

- Bin Availability Notification: Citizens can notify the app when their bin is ready for collection **during non-scheduled collection times**.
- Review and Suggestion Section: Citizens can leave reviews and suggestions to improve the service.

### 3.3 Garbage Collector Features

- Real-Time Bin Locator: Collectors can view the locations of bins placed outside in real-time on the app’s Guelma map.
- Route Optimization: The app generates the most efficient route covering all available bins.
- Route Adjustment for Blockages: **Collectors can signal if a route is blocked**, prompting the app to re-calculate a new optimal path.

### 3.4 Important Notes

Traffic Condition Exclusion: Given the lack of real-time traffic data in Guelma, route optimization will not account for traffic conditions.

## 4. Non-Functional Requirements

- User-Friendly Interface: The app should have an intuitive, accessible design for both citizens and collectors.
- Reliability: The app must provide accurate and timely information about bin locations and routes.
- Scalability: The system should be able to expand as additional neighborhoods adopt this waste management approach.

## 5. Technical Requirements

- GPS Integration: Real-time location tracking of bins and trucks.
- Map API: Integration with a mapping service to display Guelma’s layout and generate optimized routes.
- Push Notifications: Alerts for citizens on collection schedules and route updates for collectors.

## 6. Additional Features

- Blocked Route Notification: Garbage collectors can mark a route as impassable to auto-generate alternative routes.
- Feedback Section for Citizens: A space for citizens to give feedback and make suggestions for service improvements.

## 7. Project Constraints

- Traffic Data Limitation: The app will not support real-time traffic adjustments due to the unavailability of such data in Guelma.
- Hardware Requirements: Collectors will need smartphones with GPS capabilities to access the app.
