
<div dir="rtl">

## فهرس

- [I. المقدمة](#I-المقدمة)
    - [I.1 تحديد المشكلة](#I1-تحديد-المشكلة)
    - [I.2 نظرة عامة على المشروع](#I2-نظرة-عامة-على-المشروع)
- [II. أهداف المشروع](#II-أهداف-المشروع)
- [III. الميزات الرئيسية](#III-الميزات-الرئيسية)
- [IV. مجموعة التقنيات المستخدمة](#IV-مجموعة-التقنيات-المستخدمة)
    - [IV.1 تقنيات الواجهة الأمامية](#IV1-تقنيات-الواجهة-الأمامية)
    - [IV.2 تقنيات الواجهة الخلفية](#IV2-تقنيات-الواجهة-الخلفية)
    - [IV.3 تقنيات قواعد البيانات](#IV3-تقنيات-قواعد-البيانات)
- [V. الجمهور المستهدف](#V-الجمهور-المستهدف)
- [VI. الكلمات المفتاحية](#VI-الكلمات-المفتاحية)

<div>

<div dir="rtl">

## I. المقدمة

### I.1 تحديد المشكلة:
يواجه المبتدئون في البرمجة، وخاصة طلاب السنة الأولى في مجال علوم الحاسوب، تحديات في فهم الخوارزميات وغالبًا ما يخلطون بينها وبين لغات البرمجة. يشكل هذا الخلط عائقًا أمام تعلم أساسيات البرمجة بشكل فعال.

### I.2 نظرة عامة على المشروع:
TG Blocks هو منصة تهدف إلى سد الفجوة بين الخوارزميات ولغات البرمجة. تقدم هذه المنصة تمثيلًا بصريًا للخوارزميات إلى جانب الكود المقابل لها في عدة لغات برمجة. ستساعد هذه المنصة المبتدئين في فهم المنطق الكامن وراء الخوارزميات دون تعقيدات صياغة لغة البرمجة ودون تشتيت انتباههم بأخطاء المترجم.

<div>

<div dir="rtl">

## II. أهداف المشروع

- تطوير واجهة ويب سهلة الاستخدام للمنصة.  
  > الواجهة هي **الجزء الأكثر أهمية** في المشروع، لأنها أول شيء يتفاعل معه المستخدم.
- تنفيذ تمثيل مرئي للخوارزميات باستخدام البرمجة القائمة على الكتل.
- توفير الكود المقابل للخوارزمية في عدة لغات برمجة.
- تنفيذ ميزة تسمح للمستخدمين بالمساهمة في المنصة من خلال إضافة خوارزميات جديدة أو تحسين الخوارزميات الموجودة.

  > حتى الآن، لم تكن هناك نسخة جيدة من المشاريع المماثلة للأجهزة المحمولة. لذلك، سنحاول توفير نسخة للمنصة تعمل على الأجهزة المحمولة.

<div>

<div dir="rtl">

## III. الميزات الرئيسية

- **واجهة سهلة الاستخدام**: ستحتوي المنصة على واجهة بديهية وسهلة الاستخدام.
- توفير نسخة ***غير متصلة بالإنترنت*** من المنصة لأجهزة سطح المكتب.
- **تمثيل مرئي**: ستقدم المنصة تمثيلًا مرئيًا للخوارزميات باستخدام البرمجة القائمة على الكتل.
- **لغات برمجة متعددة**: ستوفر المنصة الكود المقابل للخوارزمية بعدة لغات برمجة.
    - في البداية، سنوفر لغة واحدة فقط، ويجب أن تكون لغة بسيطة ومفسرة مثل بايثون.
- **ميزة المساهمة**: ستتيح المنصة للمستخدمين المساهمة من خلال إضافة خوارزميات جديدة أو تحسين الخوارزميات الحالية.
- **نظام تسجيل الدخول**: ستحتوي المنصة على نظام تسجيل دخول يسمح للمستخدمين بحفظ تقدمهم ومساهماتهم. `غير مطلوب في النسخة غير المتصلة بالإنترنت`

<div>


<div dir="rtl">

## IV. مجموعة التقنيات المستخدمة

### IV.1 تقنيات الواجهة الأمامية:

- استخدم أي تقنيات واجهة أمامية تفضلها.  
  > نوصي باستخدام أطر التطبيقات ذات الصفحة الواحدة (SPA)، وElectron.js لتطبيقات سطح المكتب.

### IV.2 تقنيات الواجهة الخلفية:

- استخدم أي تقنيات واجهة خلفية تفضلها.

### IV.3 تقنيات قواعد البيانات:

- قاعدة بيانات SQL كقاعدة بيانات أساسية. `لإصدار الويب`

<div>

<div dir="rtl">

## V. الجمهور المستهدف:

- طلاب السنة الأولى في مجال علوم الحاسوب.
- المبتدئون في البرمجة.
- المدرسون الذين يرغبون في تعليم الخوارزميات بطريقة تفاعلية أكثر.

<div>

<div dir="rtl">

## VI. الكلمات المفتاحية:

ستساعدك هذه الكلمات المفتاحية في العثور على مزيد من المعلومات حول المشروع.

#Scratch #Algorithms #Programming #Block-Based-Programming #Electron.js


<div>
