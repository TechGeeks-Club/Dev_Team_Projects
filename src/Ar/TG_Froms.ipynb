
<div dir="rtl">
<h2> فهرس</h2>

1. [I. مقدمة](#I.-مقدمة)
    - [I.1 الخلفية ومشكلة المشروع](#I.1-الخلفية-ومشكلة-المشروع)
    - [I.2 نظرة عامة على المشروع](#I.2-نظرة-عامة-على-المشروع)
2. [II. أهداف المشروع](#II.-أهداف-المشروع)
    - [II.1 جمع البيانات وتخزينها](#II.1-جمع-البيانات-وتخزينها)
    - [II.2 مشاركة البيانات وإعادة الاستخدام](#II.2-مشاركة-البيانات-وإعادة-الاستخدام)
3. [III. الميزات الرئيسية](#III.-الميزات-الرئيسية)
    - [III.1 تطوير المنصة](#III.1-تطوير-المنصة)
      - [III.1.1 ميزات مساحة المستخدم](#III.1.1-ميزات-مساحة-المستخدم)
      - [III.1.2 ميزات مساحة المشرف](#III.1.2-ميزات-مساحة-المشرف)
    - [III.2 وظيفة البحث والاسترجاع](#III.2-وظيفة-البحث-والاسترجاع)
4. [IV. مجموعة التقنيات](#IV.-مجموعة-التقنيات)
    - [IV.1 تقنيات الواجهة الأمامية](#IV.1-تقنيات-الواجهة-الأمامية)
    - [IV.2 تقنيات الواجهة الخلفية](#IV.2-تقنيات-الواجهة-الخلفية)
    - [IV.3 تقنيات قواعد البيانات](#IV.3-تقنيات-قواعد-البيانات)

<div>

<div dir="rtl">
<h2>I. مقدمة</h2>
<h3> I.1 الخلفية ومشكلة المشروع :</h3>

في البيئة الأكاديمية الحالية، يتعين على طلاب الجامعات في السنة الثالثة والخامسة إجراء أبحاث وجمع بيانات كجزء من مشاريع التخرج الخاصة بهم. عادةً ما يعتمد الطلاب على نماذج جوجل فورمز لجمع البيانات ذات الصلة بدراستهم. ومع ذلك، بمجرد جمع البيانات واكتمال المشروع، لا يتم تخزين المعلومات ولا يتم مشاركتها لإعادة الاستخدام المحتمل. وهذا يؤدي إلى تكرار كبير للجهد حيث يجب على الطلاب الذين يعملون على مشاريع مماثلة جمع البيانات التي تم جمعها بالفعل من قبل الآخرين.

<h3> I.2 نظرة عامة على المشروع :</h3>

**ت.ج فورمز** تهدف إلى معالجة هذه المشكلة من خلال توفير منصة مركزية لطلاب الجامعات لإنشاء وإدارة وتخزين نماذج جمع البيانات. ستمكن هذه المنصة الطلاب من الوصول إلى البيانات التي تم جمعها مسبقًا والتي تكون ذات صلة بأبحاثهم، وبالتالي توفير الوقت وتقليل الجهود المتكررة وإنشاء مستودع قيم للمعلومات للاستخدام الأكاديمي المستقبلي. الهدف الطويل الأجل هو إنشاء مجموعة بيانات شاملة ستعود بالفائدة على الطلاب الذين يقومون بمشاريع مماثلة في السنوات اللاحقة.

<div>

<div dir="rtl">

<h2> II. أهداف المشروع</h2>

<h3> II.1 جمع البيانات وتخزينها :</h3>

- تطوير منصة يمكن للطلاب من خلالها إنشاء نماذج مخصصة لجمع البيانات لمشاريعهم.
- ضمان تخزين جميع البيانات المجمعة بشكل آمن في قاعدة بيانات مركزية لاسترجاعها بسهولة واستخدامها في المستقبل.

<h3> II.2 مشاركة البيانات وإعادة الاستخدام :</h3>

- السماح للطلاب بالوصول إلى البيانات التي تم جمعها مسبقًا والتي تكون ذات صلة بمواضيع أبحاثهم.

<div>

<div dir="rtl">

## III. الميزات الرئيسية 

### III.1 تطوير المنصة:
#### III.1.1 ميزات مساحة المستخدم:

- **لوحة التحكم:** يمكن للمستخدمين عرض وإدارة استماراتهم والبيانات التي تم جمعها من لوحة تحكم مركزية.
- **إنشاء الاستمارات:** يمكن للمستخدمين إنشاء استمارات مخصصة مع حقول إدخال متنوعة لجمع البيانات.
- **جمع البيانات:** يمكن للطلاب مشاركة استماراتهم مع الآخرين لجمع البيانات.
- **تخزين البيانات:** سيتم تخزين جميع البيانات المجمعة بشكل آمن في قاعدة بيانات مركزية.
- **استرجاع البيانات:** يمكن للطلاب الوصول إلى البيانات المجمعة سابقًا والتي تتعلق بمواضيع أبحاثهم.
- **تصور البيانات:** يمكن للمستخدمين عرض وتحليل البيانات المجمعة من خلال المخططات والرسوم البيانية.
- **تصدير البيانات:** يمكن للطلاب تصدير البيانات المجمعة بصيغ مختلفة لمزيد من التحليل.
- **مشاركة البيانات:** يمكن للمستخدمين مشاركة استماراتهم والبيانات المجمعة مع الآخرين للاستخدام الأكاديمي.
- **خصوصية البيانات:** ستضمن المنصة خصوصية وأمان جميع البيانات المجمعة. (يمكن للمستخدم اختيار جعل البيانات عامة أو خاصة)
- **توافق عبر المنصات:** ستكون المنصة متاحة على أجهزة متنوعة، بما في ذلك أجهزة الكمبيوتر المكتبية، وأجهزة الكمبيوتر المحمولة، والأجهزة اللوحية، والهواتف الذكية.

#### III.1.2 ميزات مساحة المشرف:

- **لوحة تحكم المشرف:** يمكن للمشرفين إدارة حسابات المستخدمين، والاستمارات، والبيانات المجمعة، وعرض الإحصائيات والتقارير.
- **إدارة المستخدمين:** يمكن للمشرفين إضافة وتحرير وحذف حسابات المستخدمين.
- **إدارة الاستمارات:** يمكن للمشرفين عرض وتحرير وحذف الاستمارات التي أنشأها المستخدمون.
- **إدارة البيانات:** يمكن للمشرفين الوصول إلى جميع البيانات المجمعة في قاعدة البيانات وإدارتها.
- **إدارة الأذونات:** يمكن للمشرفين تعيين الأدوار والأذونات للمستخدمين بناءً على مسؤولياتهم.
- **نسخ احتياطي للبيانات:** يمكن للمشرفين إجراء نسخ احتياطي واستعادة البيانات لمنع فقدان المعلومات.

### III.2 وظيفة البحث والاسترجاع:

- تنفيذ وظيفة بحث تمكن المستخدمين من تحديد مواقع البيانات ذات الصلة والوصول إليها بسرعة.
- ضمان أن تكون عملية الاسترجاع فعالة وسهلة الاستخدام.

<div>

<div dir="rtl">

## IV.  مجموعة التقنيات

### IV.1 تقنيات الواجهة الأمامية:

- استخدم أي تقنيات أمامية تفضلها. ولكن نوصي باستخدام HTML و CSS و JavaScript.

### IV.2 تقنيات الواجهة الخلفية:

- استخدم أي تقنيات خلفية تفضلها. 

### IV.3 تقنيات قواعد البيانات:

- قاعدة بيانات SQL كقاعدة البيانات الأساسية.
- وقاعدة بيانات وثائقية لتخزين الاستمارات والبيانات.

<div>
